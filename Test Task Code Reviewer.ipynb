### 1.5. Create a movie_duration_category column, which will contain three categories depending on the film length:
#### • Category "1. <90" if the film is less than 90 minutes long
#### • Category "2. 90–120" if the film is between 90 minutes and two hours long (inclusively)
#### • Category "3. >120" if the film is more than two hours long

```python
#function that returns category according to conditions
def duration_category (df_movie):
    if df_movie['duration'] < 90:
        category = 'Category 1'
    elif (df_movie['duration'] >=90) & (df_movie['duration'] <=120):
        category = 'Category 2'
    else:
        category = 'Category 3'
    return category
```

```python
#inserting column in dataframe
df_movie['movie_duration_category'] = df_movie.apply(duration_category, axis=1)
display(df_movie)
```

### 1.8. The plot_keywords column holds keywords characterizing the film's plot. Using the data in this column, create a column called movie_plot_category, to contain four categories depending on the key words in the column:
#### • Category "love_and_death" if the keywords include both "love" and "death"
#### • Category "love" if the keywords include the word "love"
#### • Category "death" if the keywords include the word "death"
#### • Category "other" if the keywords do not meet the conditions above

```python
#function that returns category according to conditions
import numpy as np

def keywords_category(df_movie):
    category = 'other'
    if pd.notnull(df_movie['plot_keywords']):  
        category_list = df_movie['plot_keywords'].split('|')
        if ('love' in category_list and 'death' in category_list):
            category = 'love_death'
        elif 'death' in category_list:
            category = 'death'
        elif 'love' in category_list:
            category = 'love'
        else:
            category = 'other'
        
    return category
```

```python
#inserting column in dataframe
df_movie['movie_plot_category'] = df_movie.apply(keywords_category, axis=1)
display(df_movie)
```

```python
#total by category
df_movie['movie_plot_category'].value_counts()
```

#### • event_type - type of event: registration - registration in the application; simple_event click event in the application; purchase - an event of purchase within the application;

```python
import pandas as pd

df_event = pd.read_csv('event_data.csv')
display(df_event)
```

### 2.2 Highlight user cohorts based on the week of registration in the application. The cohort identifier should be the week ordinal (for example, the week from July 29 to August 4 should have identifier 31).

```python
import datetime

# converting column event_date from string to date
df_event['event_date']= pd.to_datetime(df_event['event_date'])
```

```python
#inserting new column in dataframe with week ordinal
df_event['event_week_ordinal'] = df_event['event_date'].dt.isocalendar().week
df_event
```

```python
#Grouping by user_id 
grouping = df_event.groupby('user_id')['event_week_ordinal'] 
# Assigning a minimum week_ordinal value to the dataset
df_event['cohort_week_registration'] = grouping.transform('min')
# printing top 5 rows
display(df_event)
```

### 2.3 How many unique users in the cohort with ID 33?

```python
import numpy as np

#allows analyze and drop cohort_week_registration with Null values in different ways
df_event.dropna(subset=['cohort_week_registration'], inplace=True)
#groupby cohort_week_registration and unique users
num_ids = df_event.groupby(['cohort_week_registration'])['user_id'].nunique()
num_ids
```

```python
print(f'There are {num_ids[33]} unique users in the cohort with ID 33 ')
```

### 2.4. For each event, highlight the indicator lifetime - the weekly lifetime of the cohort. The lifetime indicator is calculated based on the serial number of the week in which the event is committed, relative to the week of registration. For example, an event committed on August 3 by a user from a cohort of registrants at 31 weeks will be committed on the zero week of lifetime, and an event committed by the same user on August 5 will be committed on the first week of lifetime).

```python
#calculating cohort_lifetime
df_event['cohort_lifetime'] = df_event['event_week_ordinal'] - df_event['cohort_week_registration']
df_event
```

### 2.5. Build a summary table of changes in the Retention Rate for cohorts depending on lifetime.

```python
#using pivot_table for aggregate colums in by cohort
cohorts_by_num_id = pd.pivot_table(df_event, index = ['cohort_lifetime', 'cohort_week_registration'], 
                         aggfunc = {'user_id':'nunique'}).reset_index() 
cohorts_by_num_id
```

```python
#selecting only zero cohort_lifetime values per week, indicating the beginning of the user counter
cohort_lifetime_zero = cohorts_by_num_id[cohorts_by_num_id['cohort_lifetime'] == 0]
#selecting columns
cohort_lifetime_zero = cohort_lifetime_zero[['cohort_week_registration','user_id']]

# mergind df to get the amount of user_cohort and the amount of initial user that week
df_event_cohort = cohorts_by_num_id.merge(cohort_lifetime_zero,on='cohort_week_registration')
display(df_event_cohort)
```

```python
#retention rate 
df_event_cohort['retention'] = df_event_cohort['user_id_x']/df_event_cohort['user_id_y'] 
#using pivot_table for aggregate colums in df
retention_rate = pd.pivot_table(df_event_cohort, index='cohort_week_registration',
                                      columns='cohort_lifetime',values='retention',
                                      aggfunc=np.sum) 

retention_rate
```

```python
#plot retention rate
import matplotlib.pyplot as plt
import seaborn as sns

plt.title('Retention Rate')
sns.heatmap(retention_rate.iloc[:, 1:], annot=True, fmt='.3',annot_kws={'size':12}, cbar=False);
```

### 2.6. What is the 3 week retention rate for a cohort with ID 32? Give the answer in percent, rounded to 2 decimal places, inclusive.

```python
x = (retention_rate[3][32])*100

print(f'The 3 week retention rate for a cohort with ID 32 is {x:.2f}%')
```

### 2.7. Build a summary table of changes in the indicator ARPPU (Average Revenue Per Paying User) for cohorts depending on lifetime.

```python
#filtering df by purchase
df_event_purchase = df_event[df_event['event_type'] == 'purchase']
df_event_purchase = df_event_purchase[['user_id','cohort_week_registration', 'cohort_lifetime', 'purchase_amount']]

#pivot_table with the sum of purchase per cohort
cohort_purchase = pd.pivot_table(df_event_purchase, values = 'purchase_amount',index = ['cohort_lifetime', 'cohort_week_registration'], 
                                 aggfunc=np.sum)

#pivot_table with the sum of user per cohorts
id_purchase = pd.pivot_table(df_event_purchase,index = ['cohort_lifetime', 'cohort_week_registration'], 
                                 aggfunc = {'user_id':'nunique'})
# merging dataframes 
df_ARPPU = cohort_purchase.merge(id_purchase,on=['cohort_week_registration', 'cohort_lifetime'])

# # average purchase py user
df_ARPPU['ARPPU'] = df_ARPPU['purchase_amount']/df_ARPPU['user_id']
df_ARPPU.columns = ['purchase_sum', 'users', 'ARPPU']
df_ARPPU
```

### 2. 8. What is the 3-week ARPPU of a cohort with ID 31? Give the answer with a floating point number, rounded to 2 decimal places, inclusive.

```python
x = df_ARPPU['ARPPU'][31][3]
print(f'The 3-week ARPPU of a cohort with ID 31 is {x:.2f}')
```

## "What is the difference between DataFrame and Series?"

A pandas series is a one-dimensional data structure and can only contain single list with index, it is a type of list in pandas which can take integer values, string values, double values and more. But in Pandas Series we return an object in the form of list, having index starting from 0 to n, Where n is the length of values in series. 
Follow an example:

```python
import pandas as pd

#list of country
country_list = ['Brazil', 'USA', 'Mexico', 'Canada']

#creating series
country_series = pd.Series(country_list)

print(country_series)
print(type(country_series))
```

A pandas dataframe is a two-dimensional data-structure and can be made of more than one series or we can say that a dataframe is a collection of series, also Pandas DataFrame can be created from the lists, dictionary, and from a list of dictionary etc. The Data structure contains labeled axes (rows and columns).

```python
import pandas as pd

#list of country
country_list = ['England', 'Germany', 'Netherlands', 'Portugal']

#creating dataframe
country_dataframe = pd.DataFrame(data=country_list, columns=['country'])

print(country_dataframe)
print(type(country_dataframe))
```

It is worth mentioning that series have only a top level name, dataframes have a top level and a column name. That can lead to significant differences in syntax for processing/creating a new series vs a new column and print a specific value. Follow an example:

```python
#To print the element with index 2 in a series we can use:
print(country_series[2])
#To print the element with index 2 in a dataframe it is necessary to inform the column name and the index of the row
print(country_dataframe['country'][2])
```
